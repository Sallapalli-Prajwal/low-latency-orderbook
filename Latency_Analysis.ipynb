```python
import pandas as pd, matplotlib.pyplot as plt

df = pd.read_csv("latency_samples.csv")
plt.figure(figsize=(8,5))
for tid, grp in df.groupby("thread_id"):
    grp["latency_ns"].plot(kind="kde", label=f"Thread {tid}")
plt.xlabel("Latency (ns)")
plt.ylabel("Density")
plt.legend()
plt.title("Per-thread latency distribution")
plt.show()
```
