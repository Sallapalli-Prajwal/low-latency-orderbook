```python
import pandas as pd, matplotlib.pyplot as plt
df = pd.read_csv("system_usage.csv")
fig, ax1 = plt.subplots()
ax1.plot(df["time_s"], df["rss_MB"], label="Memory (MB)", color="tab:blue")
ax2 = ax1.twinx()
ax2.plot(df["time_s"], df["cpu_s"].diff().fillna(0), label="CPU sec/sec", color="tab:red")
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("RSS (MB)")
ax2.set_ylabel("CPU usage (s/s)")
plt.title("System resource usage during stress test")
plt.show()
```
